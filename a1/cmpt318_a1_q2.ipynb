# Part 4: Linear Algebra (5 points)

Solve the given system of linear equations.
\begin{align}
x_{0}+2x_{1}\qquad = 1
\end{align}
\begin{align}
3x_{0}+2x_{1}+4x_{2} = 7
\end{align}
\begin{align}
-2x_{0}+x_{1}-2x_{2} = -1
\end{align}


HINT: AX = B, solve for X where

\begin{align}
X = [x_{0}, x_{1}, x_{2}]^{T}
\end{align}


```python
import numpy as np
from numpy.linalg import inv
# Question 4.1 (5 points): Solve for X and print the X.
# TODO: BEGIN YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)

# Equation 1: x0 + 2x1 + 0x2 = 1
# Equation 2: 3x0 + 2x1 + 4x2 = 7
# Equation 3: -2x0 + x1 - 2x2 = -1
A = np.array([
    [1, 2, 0],
    [3, 2, 4],
    [-2, 1, -2]
])

# Result maxtrix
B = np.array([1, 7, -1])

# Compute the inverse of A and solve for X
A_inverse = inv(A)
X = np.dot(A_inverse, B)

print(f"X: {X}")

# END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)
```
